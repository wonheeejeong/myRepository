# 1. 서론




*   온라인 게임, 특히 MMORPG(Massively Multiplayer Online Role-Playing Game)는 **수많은 유저들이 인터넷을 통해 주어진 역할을 수행하는 게임**을 말한다.

*   유저들은 게임공간 안에서 **몬스터를 사냥해서 전리품과 화폐를 획득하며, 이를 다른 사람들과 같이 거래할 수 있다.**

*   게임개발사는 유저들의 플레이방식을 분석하여 게임경제가 적절하게 유지되도록 **게임화폐가치 및 게임재화가치를 조절하는 역할을 수행**한다.

*   일반적으로 MMORPG는** '거래소'라는 시스템이 있어서 유저들은 일정 수수료를 지불하고 자유롭게 자신의 아이템을 거래할 수 있다.**

*   이와 다르게 게임 내 시장이 아닌** 게임 외적인 공간에서 아이템을 거래하는 시장이 존재하는데, '아이템 거래 사이트'라고 한다.**

***게임경제와 실물경제의 차이점***

■   게임경제는 최저생계비가 필요하지 않다.

>* 실물경제에서 사람들은 기본적인 최저생계비가 필요하다(주거비용, 식비 등) 

>* 하지만 게임경제에서는 최저생계비가 없기때문에 **게임개발사가 따로 화폐 회수 시스템을
구축하지 않으면 필연적으로 인플레이션이 발생할 수 밖에 없다.**

>* 따라서 게임개발사는 게임경제 및 인플레이션 관리를 위한 여러가지 화폐 회수 컨텐츠를 개발하고 관리한다.(e.g 장비 수리비, 소모품 비용, 강화비용 등)

■   게임경제는 PvE(Player versus Environment)가 주된 컨텐츠이다.

>* MMORPG는 게임서버에 설계된 게임환경에 의해 게임이 진행된다.

>* 필드에 생성되는 몬스터를 죽여서 아이템을 얻거나, 값비싼 전리품을 얻기 위해 위험한 던전에 들어가기도 한다.

>* PvP(Player versus Player)를 통해 보상을 얻기도 하지만 일반적으로 비주류 컨텐츠이다.

>* 게임 내 PvE의 비중이 커야 게임 개발자가 게임시스템을 통제하기 쉽기 때문이다.

>* 따라서 실물경제에 비해서 게임경제는 통제할 수 있는 범위가 굉장히 넓은 편이다.

■   치명적인 오류나 사고가 발생했을 때 유연하게 대처가 가능하다.

>* 게임 내 버그나 시스템 오류로 인한 피해가 발생했을 때 게임서버의 로그를 분석해서 문제 해결이 가능하다.

>* 하지만 실물경제는 재난 상황이나 국제경제의 영향에 의한 피해에 대한 민감도가 높다.


# 2. 가설 설정
* 게임 아이템을 거래할 수 있는 시장은 두 곳이 있다.

* 하나는 게임 내 시장인 '거래소'이고, 나머지는 게임 외 시장인 '아이템 거래 사이트'이다.

* **'거래소'는 게임 내에 존재하고 매우 손쉽게 아이템을 사고 팔 수 있는 장벽이 낮은 시장**이다.

*  따라서 '거래소'는 효율적인 시장으로 게임 아이템의 가격에 영향을 줄만한 정보(게임 업데이트 예정 공시, 아이템 밸런스 패치 등 )가 매우 빠르게 반영된다.

* 이에 비해서 '아이템 거래 사이트'는 거래하기 위한 절차가 까다롭고 복잡하여 장벽이 높은 시장이다.

* 즉, **'아이템 거래 사이트'는 가격에 대한 정보의 반영이 느린 시장**이다.

* 그런데 **두 시장 모두 같은 재화에 대한 거래가 이루어 지는 시장**이다.


◎  **정보 반영 속도 차이로 인해 '거래소'가 '아이템 거래 사이트'의 선행지표가 될 수 있다.**

▼ 일별 게임 아이템 거래가격

* 유저들이 가장 많이 거래하는 재료아이템 두 가지(마그토늄, 데이터 칩)를 선정하였다.

* 두 아이템은 MMORPG의 주요 컨텐츠인 레이드(다수의 유저들이 강력한 몬스터를 잡기 위한 협동 임무)입장 시 필요한 아이템으로, 거래소에서 거래량이 가장 많은 아이템이다.

* 위 두 아이템의 거래소 데이터를 api로 불러와서 거래량과 가격을 기록한다. 

* 과거부터 이 데이터를 모아서 일별 게임 아이템 거래가격을 산출한다.

```
# API를 활용한 던전앤파이터 거래소 아이템 데이터 크롤링   (마그토늄)


import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import types
from types import *


url = 'https://api.neople.co.kr/df/auction?limit=400&itemName=%eb%a7%88%ea%b7%b8%ed%86%a0%eb%8a%84&sort=unitPrice:asc&limit=%3Climit%3E&wordType=match&apikey=JBIBYEzdSbwZF3k4QQRWcM5jnhDwCYSn'  #오픈api를 통해 거래소 데이터 요청
req = requests.get(url)
html = req.text
soup = BeautifulSoup(html, 'html.parser')



def is_start(num):
  while html[num] not in '1234567890':
    num += 1
  return num

def is_end(num):
  while html[num] != ',':
    num += 1
  return num


Date = []
Vol = []
Price = []

check = 0


while html.find("auctionNo") != -1: # 경매장 맨 마지막 리스트까지 검색
  date = html.find("regDate") # 경매장에 아이템을 등록하는 시기
  t = str(html[(date+10):(date+29)])
  Date.append(t)


  count = html.find("count") # 등록된 아이템의 개수
  v = int(html[is_start(count):is_end(count)])
  Vol.append(v)


  price = html.find("unitPrice") # 등록된 아이템의 개당 가격
  p = int(html[is_start(price):is_end(price)])
  Price.append(p)



  check = int(price+10)

  html = html[check:]



df = { # Date, Volum, UnitPrice로 데이터 정렬
    'Date' : Date,
    'Volume' : Vol,
    'UnitPrice' : Price
}

dataFrame = pd.DataFrame(df)
dataFrame
```

* 위 방식과 같이 오픈 api를 통해 과거부터 거래소 데이터를 받아서 아이템의 거래가격을 도출하고 싶었다.

* 하지만 **문의 결과 오픈 api는 '현재 거래소 데이터'만 제공**하기 때문에 현재부터 데이터를 축적하는 방법밖에 없다.

# 4-1. 시나리오1

y = mprice(t) + dprice(t)의 경우

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
import statsmodels.stats.sandwich_covariance as sw
import statsmodels.stats.api as sms
import statsmodels.stats.stattools as smt
from statsmodels.compat import lzip
from arch import arch_model


set1 = { 
    'Date' : chipPrice.loc[:,'Date'],
    'goldPrice': goldPrice.loc[0:34,'Price'],
    'magPrice' : magPrice.loc[:,'Price'],
    'chipPrice' : chipPrice.loc[:,'Price'] 
}

data1 = pd.DataFrame(set1)

data1
```

```
result1 = sm.ols(formula = 'goldPrice ~ magPrice + chipPrice', data = data1).fit()

result1.summary()
```

**결과요약**

* magPrice의 p-value는 0.182로 설명력이 없었다.
